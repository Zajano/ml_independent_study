### **Finding the optimum number of clusters for K-Means and determining the value of K.**

```
from sklearn.feature_selection import f_regression, mutual_info_regression

# new seed for testing each time 
np.random.seed()

# 1000 items with 3 random 'feature' values
X = np.random.rand(1000, 3)

# linear for feature 1
a = X[:, 0]

# coupled for feture 2
b = np.cos(6 * np.pi * X[:, 1])

# entirely different calc for feature 3
c = np.random.rand(1000)

# target to predict
y = a + b + c
```

```
# feature score values
f_test, p_vals = f_regression(X, y)
mi_test = mutual_info_regression(X, y)

# scale to 1 for normalized values
f_test /= np.max(f_test)
mi_test /= np.max(mi_test)

# plot y against each feature to see impact
plt.figure(figsize=(15, 5))
for i in range(3):
    plt.subplot(1, 3, i + 1)
    plt.scatter(X[:, i], y, edgecolor='black', s=20)
    plt.xlabel("X_" + str(i), fontsize=12)
    if i == 0:
        plt.ylabel("y", fontsize=12)
    plt.title("f_reg=" + str(round(f_test[i], 2)) + "    mi_reg=" + str(round(mi_test[i], 2)),
              fontsize=18)

plt.savefig("feature importance.png", bbox_inches='tight')
plt.show()
```
